```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.formula.api import ols
import patsy
from plotnine import *
from scipy.stats import ttest_ind
```

```python
# Anti Join Function Used Later
def anti_join(x, y, on):
    """Return rows in x which are not present in y"""
    ans = pd.merge(left=x, right=y, how='left', indicator=True, on=on)
    ans = ans.loc[ans._merge == 'left_only', :].drop(columns='_merge')
    return ans

def anti_join_all_cols(x, y):
    """Return rows in x which are not present in y"""
    assert set(x.columns.values) == set(y.columns.values)
    return anti_join(x, y, x.columns.tolist())
```

```python
# Reading in the data
analysis_df = pd.read_csv('../20_intermediate/confounding_factors.csv')

# Deleting two extra columns
del analysis_df['Unnamed: 0']
del analysis_df['Unnamed: 0.1']

# Some quick analysis
analysis_df.describe()
```

### Some Initial EDA

```python
# Intensity Does have an effect on percent population change.
ttest_ind(analysis_df['Pop_percent_change'], analysis_df['intensity'], equal_var=True)
```

```python
# Histogram
sns.distplot(analysis_df['Pop_percent_change'], color='green')
```

```python
sns.jointplot('Pop_percent_change', 'estimated_yield', data=analysis_df, color='blue')
```

```python
# Filtering the treatment to > 1 fatality per 10k and after 2013. This number based on previous analysis of 2014 and 2015 migration paterns
analysis_df['violent'] = np.where((analysis_df['intensity'] > 1) & (analysis_df['Year'] >=2014), 1, 0)
```

```python
# Building a df with the violent counties after 2014
violent_df = analysis_df[analysis_df['violent'] == 1]

# Saving the violent counties into a list
violent_counties = violent_df['County'].unique()

# Creating a DF with violent counties for the whole time period (2011 - 2015)
violent_analysis_df = analysis_df[analysis_df['County'].isin(violent_counties)]
```

```python
# Non Violent DF
non_violent_analysis_df = anti_join_all_cols(analysis_df, violent_analysis_df)
```

```python
print("The violent DF dimensions are: ", violent_analysis_df.shape)
print("The non violent DF dimensions are: ", non_violent_analysis_df.shape)
print("Which sums up to the original DF: ", analysis_df.shape)
```

```python
###############################################
#### Pre Post Plot Prep ####
###############################################
#Sorting the data
analysis_df_sorted = analysis_df.sort_values('Year')

# Grouping by year
df_grouped = analysis_df_sorted.groupby('Year').mean() 

# Reindexing
df_grouped.reset_index(inplace=True)

###############################################
#### Difference and Difference Plot Prep ####
###############################################
#Sorting the data
high_violence_df_s = violent_analysis_df.sort_values('Year')
low_violence_df_s = non_violent_analysis_df.sort_values('Year')

# Grouping by year
hv = high_violence_df_s.groupby('Year').mean() 
lv = low_violence_df_s.groupby('Year').mean() 

# Reindexing
hv.reset_index(inplace=True)
lv.reset_index(inplace=True)
```

```python
##################
### Plotting  ####
##################

fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (20,12))
axes[0].plot('Year','Pop_percent_change', data = hv, label= 'High Violence Counties', color='red', ls='-')
axes[0].plot('Year', 'Pop_percent_change', label = 'Low Violence Counties', data = lv, ls ='--')
axes[0].axvline(2011, -3.3, 4.5, label='Country Founded', color = 'orange')
axes[0].axvline(2013.92, -3.3, 4.5, label='Conflict Start', color = 'black', ls=":")
axes[0].legend()
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Percent Population Change')
axes[0].set_title('Difference and Difference Analysis')

axes[1].plot('Year','Pop_percent_change', data = df_grouped, label= 'Percentage Population Trend', color='grey', ls='-')
axes[1].axvline(2011, -3.3, 4.5, label='Country Founded', color = 'orange')
axes[1].axvline(2013.92, -3.3, 4.5, label='Conflict Start', color = 'black', ls=":")
axes[1].legend()
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Percent Population Change')
axes[1].set_title('Pre Post Analysis')
print('Note: Since we are only look at whether or not people leave, not where they go. Its entirely possible that IDPs go from a high violence county to a low violence one')
print('\n')
```

# Modeling

## Violent vs Not Violent County analysis

```python
# Ensuring every observation in the violence DF did was labeled
violent_analysis_df['violent'] = 1
```

```python
# Creating a list for concatination
frames = [violent_analysis_df, non_violent_analysis_df]

# Concatinating 
violence_level_df = pd.concat(frames)

# Validating the correct observations (280/280)
violence_level_df['violent'].sum()
```

```python
violence_level_df.head(1)
```

```python
# Starting with Statsmodels (No state or county controls)
variables_0 = 'Pop_percent_change ~ C(violent)'
smf.ols(variables_0, data = violence_level_df).fit().summary()
```

```python
violence_level_df.head(1)
```

```python
# Starting with Statsmodels (No state, county,  or intensity controls)
variables_0 = 'Pop_percent_change ~ C(violent) + Hospitals + estimated_yield + Population_Percent_Child + Population_Percent_Female + Poverty_Rate + C(Airport)'
smf.ols(variables_0, data = violence_level_df).fit().summary()
```

```python
# Starting with Statsmodels (No state, county,  or intensity controls)
variables_0 = 'Pop_percent_change ~ C(violent) + Population_Percent_Child'
smf.ols(variables_0, data = violence_level_df).fit().summary()
```

## Changing treatment to impending violence

For this section I decided to see if the violence the following year would effect population change in the current year.
The idea being that people would flee in anticipation of violence. Also we did not analyze 2015 because we did not have next year data on it.

```python
# analysis_df.to_excel('analysis_df_w_next_year_intensity.xlsx')
# Reading in the Excel after creating the column
next_year_intensity_df = pd.read_excel('../20_intermediate/analysis_df_w_next_year_intensity.xlsx')

# Filtering out 2015 
next_year_intensity_df = next_year_intensity_df[next_year_intensity_df['Year'] < 2015]

# Printing the head
next_year_intensity_df.head(1)
```
