```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.formula.api import ols
import patsy
from plotnine import *
from scipy.stats import ttest_ind
```

```python
# Reading in the data
analysis_df = pd.read_csv('../20_intermediate/confounding_factors.csv')

# Deleting two extra columns
del analysis_df['Unnamed: 0']
del analysis_df['Unnamed: 0.1']

# Some quick analysis
analysis_df.describe()
```

```python
analysis_df.head(3)
```

#### Note: mean Population Percent Change is 4%. Additionally it might make sense to multiply Population_Percent_Child, Population_Percent_Female, and Poverty_Rate by 100 to get more explainable coefficients

```python
# Filtering the treatment to > 1 fatality per 10k and after 2013. This number based on previous analysis of 2014 and 2015 migration paterns
analysis_df['violent'] = np.where((analysis_df['intensity'] > 1) & (analysis_df['Year'] >=2014), 1, 0)
```

```python
###############################################
#### Pre Post Plot Prep ####
###############################################
#Sorting the data
analysis_df_sorted = analysis_df.sort_values('Year')

# Grouping by year
df_grouped = analysis_df_sorted.groupby('Year').mean() 

# Reindexing
df_grouped.reset_index(inplace=True)

###############################################
#### Difference and Difference Plot Prep ####
###############################################
# Subsetting the DFs (for plotting)
low_violence_df = analysis_df[analysis_df['violent'] == 0]
high_violence_df = analysis_df[analysis_df['violent'] == 1]

#Sorting the data
high_violence_df_s = high_violence_df.sort_values('Year')
low_violence_df_s = low_violence_df.sort_values('Year')

# Grouping by year
hv = high_violence_df_s.groupby('Year').mean() 
lv = low_violence_df.groupby('Year').mean() 

# Reindexing
hv.reset_index(inplace=True)
lv.reset_index(inplace=True)
```

```python
##################
### Plotting  ####
##################

fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (20,12))
axes[0].plot('Year','Pop_percent_change', data = hv, label= 'High Violence Counties', color='red', ls='-')
axes[0].plot('Year', 'Pop_percent_change', label = 'Low Violence Counties', data = lv, ls ='--')
axes[0].axvline(2011, -3.3, 4.5, label='Country Founded', color = 'orange')
axes[0].axvline(2013.92, -3.3, 4.5, label='Conflict Start', color = 'black', ls=":")
axes[0].legend()
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Percent Population Change')
axes[0].set_title('Difference and Difference Analysis')

axes[1].plot('Year','Pop_percent_change', data = df_grouped, label= 'Percentage Population Trend', color='grey', ls='-')
axes[1].axvline(2011, -3.3, 4.5, label='Country Founded', color = 'orange')
axes[1].axvline(2013.92, -3.3, 4.5, label='Conflict Start', color = 'black', ls=":")
axes[1].legend()
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Percent Population Change')
axes[1].set_title('Pre Post Analysis')
print('Note: Since we are only look at whether or not people leave, not where they go. Its entirely possible that IDPs go from a high violence county to a low violence one')
print('\n')
```

# Modeling

## Difference and Difference
https://www.unifyingdatascience.org/html/interpreting_indicator_vars.html </br>
The coefficient is the difference with the other controls.

```python
# Just keeping the violence threshold because of multicolinearity
variables_0 = 'Pop_percent_change ~ C(violent)'
smf.ols(variables_0, data = analysis_df).fit().summary()
```

```python
# Adding some interaction terms 
# This was the best combinationg of explainability and accuracy I found
variables_f2 = 'Pop_percent_change ~ C(State) + C(violent) + Population_Percent_Child*C(violent) + Population_Percent_Female + Poverty_Rate'
smf.ols(variables_f2, data = analysis_df).fit().summary()
```

## Pre Post (Conflict Starting in 2014)

```python
# Adding a Conflict Start Indicator
analysis_df['conflict_start'] = np.where(analysis_df['Year']>= 2014, 1, 0)
analysis_df.head(1)
```

```python
# Starting with Statsmodels
# The Conflict starting does have a significant impact
variables_0 = 'Pop_percent_change ~ C(conflict_start)'
smf.ols(variables_0, data = analysis_df).fit().summary()
```

```python
# Started with the Conflict and the State, and progressively added more  + C(violent)
variables_1 = 'Pop_percent_change ~ C(State) + C(conflict_start) + Population_Percent_Child + Population_Percent_Female + Poverty_Rate'
smf.ols(variables_1, data = analysis_df).fit().summary()
```
